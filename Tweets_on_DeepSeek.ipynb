```python
from deep_translator import GoogleTranslator
def translate_text(text):
    try:
        translated=GoogleTranslator(source='auto', target='en').translate(text)
        return translated
    except:
        return text
```

Finding Outlayers:-

```python
plt.boxplot(x=Df['likeCount'])
plt.show()
```

```python
Df_Sample = Df.head(50)
Df_Sample['text'] = Df_Sample['text'].apply(translate_text)
```

```python
Df_Sample
```

```python
keywords=['Deepseek','#Deepseek','DeepSeekTech','DeepSeekAI','DeepSeek vs Claude','Liang Wenfeng','DedalusDev']
```

```python
pattern = '|'.join(keywords)
Filtered_Df_Sample=Df_Sample[Df_Sample['text'].astype(str).str.contains(pattern,  case=False)]
```

```python
Filtered_Df_Sample
```

Analyzing with plots:-

```python
lang_Counts=Filtered_Df_Sample['lang'].value_counts()

plt.bar(x=lang_Counts.index,height=lang_Counts.values,data=Filtered_Df_Sample,color='black',alpha=0.5)
plt.xlabel('Tweet Languages')
plt.ylabel('Language Count')
plt.title('Number of Tweets in Different languages')
plt.show()
```

```python
sns.countplot(x=Filtered_Df_Sample['isReply'],data=Filtered_Df_Sample)
plt.xlabel('Tweet is a Reply/Not')
plt.title('Count of replyed Tweets')
plt.show()
```

```python
plt.hist(Filtered_Df_Sample["viewCount"], bins=10, edgecolor="black", alpha=0.7)
plt.xlabel("View Count")
plt.ylabel("Frequency")
plt.title("Distribution of View Counts")

plt.show()
```

```python
Filtered_Df_Sample["viewCount"].mean()
```

```python
sns.catplot(data=Filtered_Df_Sample, x=Filtered_Df_Sample['author_isBlueVerified'],y=Filtered_Df_Sample['author_following'], hue=Filtered_Df_Sample['author_isBlueVerified'], kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10, 5))  
plt.scatter(x=Filtered_Df_Sample['pseudo_id'],y=Filtered_Df_Sample['author_followers'],)
plt.xticks(ticks=Filtered_Df_Sample['pseudo_id'],labels=Filtered_Df_Sample['pseudo_id'], rotation=45)
plt.xlabel('Pseudo ID')  
plt.ylabel('Author Followers')  
plt.title('Scatter Plot of Author Followers vs Pseudo ID')  
plt.show()
```

```python
sns.jointplot(x=Filtered_Df_Sample['viewCount'],y=Filtered_Df_Sample['likeCount'],color='orange')
plt.show()
```

```python
sns.violinplot(x=Filtered_Df_Sample['author_isBlueVerified'],  y=Filtered_Df_Sample['viewCount'])
plt.show()
```

```python
sns.catplot(x=Filtered_Df_Sample['lang'], y=Filtered_Df_Sample['viewCount'], kind="strip")
```

```python
corr_matrix = Filtered_Df_Sample[['retweetCount','author_followers', 'viewCount','likeCount']].corr()
plt.figure(figsize=(9, 6))
sns.heatmap(corr_matrix, annot=True,cmap='cividis',fmt=".2f")
plt.title('Correlation Heatmap of Tweet Engagement')
plt.show()
```

Conclusion:-

1.By analyzing the above dataset, we observe that 22% of the tweets among all collected tweets were made on Twitter related to DeepSeek.
2.The majority of the tweets are in English, indicating that DeepSeek has gained significant global attention in a short span of time.
3.Most of the tweets were posted by non-verified Twitter accounts. This suggests that the topic has garnered interest primarily from the general public rather than verified influential figures or organizations.
4.The highest number of view counts is 35, indicating the peak engagement observed in the dataset.
5.The average number of likes per tweet is low, suggesting that while the topic is widely viewed, it may not be generating significant interaction.
6.Retweet activity is minimal, showing that the majority of users are consuming the information rather than sharing it.
7.Geographical insights reveal that tweets are predominantly from countries like the USA, India, and Turkey, showcasing global interest across different regions.
8.View counts are higher for non-verified users, indicating that general accounts have shown more interest in DeepSeek compared to verified users.
9.There is a positive correlation between like count, retweet count, and view count, suggesting that tweets with higher engagement are more likely to be retweeted and viewed.
10.The positive correlation between View Count and Retweet Count indicates that retweets significantly contribute to the widespread reach of DeepSeek-related content. 
    
    This rapid spread highlights how DeepSeek has gained attention within a short period, establishing itself as a strong competitor in its domain.
